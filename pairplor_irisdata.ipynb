```python
#importing the dataset
import seaborn as sns 
iris = sns.load_dataset("iris")
iris.head()
```

```python
#plotting a default pairplot
sns.pairplot(iris)
```

## plotting with categorical features

```python
#plotting with categorical features
sns.pairplot(iris, hue="species")
```

```python
#changing the palette to "Set2"
sns.pairplot(iris, hue="species", palette = "Set2")
```

```python
#changing the markers for each level of hue variable
sns.pairplot(iris, hue="species", palette = "Set2", markers=["o", "s", "D"])
```

```python
#changing the height and aspects
sns.pairplot(iris, hue="species", palette = "Set2", size=3.5, aspect=4)
```

## customizing the plots

```python
#changing the plot type
sns.pairplot(iris,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "Set2")
```

```python
#changing the transparency, size, and edgecolor of the scatterplot dots 
sns.pairplot(iris,hue = 'species',diag_kind = "kde",plot_kws = {'alpha': 0.7, 's': 100, 'edgecolor': 'k'}, kind = "scatter",palette = "Set2")
```

```python
#plotting a subset of data
sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
```

```python
#Plotting subset of data in rows and columns
sns.pairplot(iris,x_vars=["sepal_width", "sepal_length"],y_vars=["petal_width", "petal_length"], size=3)
```

```python
#fitting linear regression
sns.pairplot(iris, kind="reg")
```
